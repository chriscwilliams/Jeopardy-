```python
import pandas as pd

# Importing CSV file
dataset = pd.read_csv("jeopardy.csv")

dataset
```

```python
def word_filter(dataset, words):
    filter = lambda x: all(word.lower() in x.lower() for word in words)
    return dataset.loc[dataset[" Question"].apply(filter)]

word_filter(dataset, ["Leap", "Day"])
```
